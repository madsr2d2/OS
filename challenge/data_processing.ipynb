# Extended_server

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# REP_PROB_PERCENT values
rep_prob_percent = [0, 20, 40, 60, 80]

# Dataset 1
mean_waiting_time_1 = [2.16356e+08, 1.88136e+08, 2.12338e+08, 2.53396e+08, 2.43668e+08]
standard_error_1 = [5.37511e+06, 4.97169e+06, 5.32578e+06, 6.82269e+06, 6.68676e+06]

# Dataset 2
mean_waiting_time_2 = [2.44274e+08, 2.40072e+08, 2.56114e+08, 2.84542e+08, 2.65383e+08]
standard_error_2 = [6.50183e+06, 6.20254e+06, 6.53506e+06, 7.41219e+06, 7.27285e+06]

# Dataset 3
mean_waiting_time_3 = [1.71125e+08, 1.89068e+08, 2.05802e+08, 2.33592e+08, 2.14962e+08]
standard_error_3 = [4.73151e+06, 4.92489e+06, 5.24541e+06, 6.05387e+06, 5.90178e+06]

# Task 1: Calculate combined mean using weights
weights_1 = 1 / np.square(standard_error_1)
weights_2 = 1 / np.square(standard_error_2)
weights_3 = 1 / np.square(standard_error_3)
combined_mean_v0 = (np.multiply(mean_waiting_time_1, weights_1) + 
                 np.multiply(mean_waiting_time_2, weights_2) + 
                 np.multiply(mean_waiting_time_3, weights_3)) / (weights_1 + weights_2 + weights_3)

# Task 2: Calculate new standard errors for combined mean
mean_waiting_times_v0 = np.array([mean_waiting_time_1, mean_waiting_time_2, mean_waiting_time_3])
combined_stdev_v0 = np.std(mean_waiting_times_v0, axis=0)
combined_sem_v0 = combined_stdev_v0 / np.sqrt(3)  # Dividing by the square root of number of observations (3)

# Task 3: Plot the combined mean and standard error
plt.figure(figsize=(10, 6))
plt.errorbar(rep_prob_percent, combined_mean_v0, yerr=combined_sem_v0, fmt='o', ecolor='black', capsize=5, label='extended_server')

# Add labels, title, and legend
plt.xlabel('Repetition Probability Percent (REP_PROB_PERCENT)')
plt.ylabel('Mean Waiting Time')
plt.title('Combined Mean Waiting Time vs REP_PROB_PERCENT')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
